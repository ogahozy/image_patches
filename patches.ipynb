```python
# import the needed libraries
import os
import sys
import shutil    # a python method that copy content from source file to  destination file. it also preserve permsn mode
import glob      # to retrieve files/pathnames matching a specified pattern
#import matplotlib.pyplot as plt
#import matplotlib.image as mpimg
from PIL import Image
```

```python
# create a directory and delete the content of existing one
def create_dir(path):
    if (os.path.exists(path)) and (os.listdir(path) != []): # if path exit and it is not empty
        shutil.rmtree(path) # used to delete an entire directory tree
        os.makedirs(path) # make a new directory
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
#crop function of the original image
def crop(input_file, height, width): # image location name, patch height and width
    img = Image.open(input_file) # read in the image using pillow
    img_width, img_height = img.size # assign the size to width and height
    for i in range(img_height//height): # normal floor division but returns highest integer
        for j in range(img_width//width):
            box = (j*width, i*height, (j+1)*width, (i+1)*height)
            yield img.crop(box) # yield is keyword that returns a generator. read up what is generator
```

```python
# split function for the new image
def split(inp_img_dir, out_dir, height, width, start_num):
    
    image_dir = os.path.join(out_dir)
    create_dir(out_dir)
    create_dir(image_dir)
    
    img_list = [f for f in os.listdir(inp_img_dir) if os.path.isfile(os.path.join(inp_img_dir, f))]
    
    file_num = 0
    for infile in img_list:
        infile_path = os.path.join(inp_img_dir, infile)
        for k, piece in enumerate(crop(infile_path, height, width), start_num):
            img = Image.new('RGB', (height, width), 255)
            img.paste(piece)
            img_path = os.path.join(image_dir, infile.split('.')[0] + '_'+ str(k).zfill(2) + '.tif')
            img.save(img_path)
 
        file_num += 1                             
        #print("\rFile %s was processed." % file_num)
        sys.stdout.write("\rFile %s was processed." % file_num)
        sys.stdout.flush()
```

```python
inp_img_dir = './test'
out_dir = './test2'
height = 512
width = 512
start_num = 1
```
